```python
# %% General Modules
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.read_excel('Glomeruli_combined.xlsx', sheet_name='in')
data.head()
```

```python
data = data.rename( columns={'Var1': 'bag', 'Var2': 'inst', 'Var3': 'img', 'ESRD': 'label'} )
data['label']=data['label'].map({0:'CTR',1:'CASE'})
data.head()
```

```python
# data = pd.read_csv('inst.csv')
# data.head()
```

```python
data_feature = data.loc[:, data.columns.str.startswith('feat')]
data_feature.head()
```

```python
data_non_feat = data.loc[:, ~data.columns.str.startswith('feat')]
data_non_feat.head()
```

```python
x = data_feature.values
```

```python
x = StandardScaler().fit_transform(x)
```

```python
pca = PCA(n_components=50)
```

```python
principalComponents = pca.fit_transform(x)
```

```python
print(pca.explained_variance_ratio_, sum(pca.explained_variance_ratio_))
```

```python
column_name = [str('principal component ')+ str(i) for i in range(1,51)]
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = column_name)
```

```python
finaldf = pd.concat([data_non_feat, principalDf], axis = 1)
```

```python
finaldf.shape
```

```python
finaldf.head(10)
```

```python
finaldf.to_csv('pca_inst.csv', index = False)
```
